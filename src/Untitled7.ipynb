```python
size = 200
nx = 100
R = 64
T = np.linspace(-np.pi,np.pi,nx)
V1 = []
V2 = []
for k in range(0,nx):
    x = R * np.cos(T[k])
    y = R * np.sin(T[k])
    proj = vp.Projector(size = size)

    proj.addObject(0,0,0)
    proj.addObject(x,y,0)
    proj.rotateObject(proj.listObjects[0],0,0,0)
    proj.computeAllVisualField()
    proj.derivateAllVisualField()
    V1.append(np.sum(proj.allVisualField[:,:,0]))
    V2.append(np.sum(proj.allVisualField[:,:,1]))
```

```python
plt.plot(V1,'o')
plt.plot(V2,'o')
```

```python
size = 200
nx = 100
R = 64
nxx = np.linspace(16,3500,10).astype(np.int)
T = []
for nx in nxx:
    t0 = time.time()
    x = R * np.cos(0)
    y = R * np.sin(0)
    proj = vp.Projector(size = nx)

    proj.addObject(0,0,0)
    proj.addObject(x,y,0)
    proj.rotateObject(proj.listObjects[0],0,0,0)
    proj.computeAllVisualField()
    proj.derivateAllVisualField()
    T.append(time.time()-t0)
```
