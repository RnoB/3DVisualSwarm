```python
scale = 2**np.arange(4,14)
```

```python
integr = []
visu = []
cont = []
sine = []
w= -1
for k in range(0,6):
    integr.append([])
for size in scale:
    
    simulator = vs.Simulator(size = size,dim = 2,parametersV =np.array([[2,.05,0],[2,.05,0],[0,0,0]]))
    simulator.scene.positionObject(0,(0,0,0))
    simulator.scene.positionObject(1,(20,0,0))
    simulator.scene.rotationObject(0,(0,0,0))
    simulator.scene.rotationObject(1,(0,0,0))
    simulator.scene.computeAllVisualField()
    simulator.scene.derivateAllVisualField()
    simulator.integrator()
    for k in range(0,6):
        integr[k].append(simulator.vIntegral[:,k])
    visu.append(simulator.scene.allVisualField)
    cont.append(simulator.scene.allVisualFieldContour)
    sine.append(simulator.scene.sine)
for j in range(0,6):
    integr[j] = np.array(integr[j])    
```
